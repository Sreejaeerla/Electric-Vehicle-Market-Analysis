# Data Cleaning

```python
import pandas as pd
import re
```

```python
df = pd.read_csv("EV Database2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
round((df.isnull().sum()/len(df))*100,2)
```

```python
df["germany_price_euros"] = df["germany_price_euros"].fillna(df["germany_price_euros"].mode()[0])
```

```python
df["netherlands_price_euros"] = df["netherlands_price_euros"].fillna(df["netherlands_price_euros"].mode()[0])
```

```python
df["uk_gbp"] = df["uk_gbp"].fillna(df["uk_gbp"].mode()[0])
```

```python
df.info()
```

```python
df1 = pd.DataFrame()
df1
```

```python
df1["brand_name"] = df["name"].apply(
    lambda x: re.match(r'^([^\s]+)\s+(.*)$', str(x)).group(1)
)
```

```python
df1["model"] = df["name"].apply(
    lambda x: re.match(r'^([^\s]+)\s+(.*)$', str(x)).group(2)
)
```

```python
df1["efficiency(Wh/km)"] = df["efficiency"].apply(lambda x : re.findall(r"\d+",x)[0]).astype("int")
```

```python
df1["real_Range(km)"] = df["real_range"].apply(lambda x : re.findall(r"\d+",x)[0]).astype("int")
```

```python
df1["weight(kg)"] = df["weight"].apply(lambda x : re.findall(r"[\d,]+", x)[0].replace(",", "")).astype("int")
```

```python
df1["acceleration(sec)"]= df["acceleration"].apply(lambda x : re.findall(r'[\d.]+',x)[0]).astype("float")
```

```python
df1["one_stop_range(km)"] = df["one_stop_range"].apply(lambda x : re.findall(r"\d+", x)[0]).astype("int")
```

```python
df1["one_stop_range(km)"] = df["one_stop_range"].apply(lambda x : re.findall(r"\d+", x)[0]).astype("int")
```

```python
df1["fast_charge(kWh)"] = df["fast_charge"].apply(lambda x : re.findall(r"[\d.]+", x)[0]).astype("float")
```

```python
df["tow_weight"] = df["tow_weight"].replace(["unknown"],"0")
```

```python
df1["tow_weight(kg)"] = df["tow_weight"].apply(lambda x : re.findall(r"[\d,]+", x)[0].replace(",", "")).astype("int")
```

```python
df["cargo_volume"] = df["cargo_volume"].replace(["unknown"],"0")
```

```python
df1["cargo_volume(L)"] = df["cargo_volume"].apply(lambda x : re.findall(r"\d+",x)[0]).astype("int")
```

```python
df1.head()
```

```python
df["price_per_range "].unique()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df1['price_per_range(€/km)']  = df["price_per_range"].apply(lambda x : re.findall(r"\d+",x)[0]).astype("int")
```

```python
df1["germany_price(euros)"] = df["germany_price_euros"].apply(lambda x : re.findall(r"[\d,]+",x)[0].replace(",","")).astype(int)
```

```python
df1["netherlands_price_(€)"] = df["netherlands_price_euros"].apply(lambda x : re.findall(r"[\d,]+",x)[0].replace(",","")).astype(int)
```

```python
df1["uk_gbp"] = df["uk_gbp"].apply(lambda x : re.findall(r"[\d,]+",x)[0].replace(",","")).astype(int)
```

```python
df1['availability'] = df['availability'].apply(
    lambda x: re.findall(r'^[^\s]+', str(x))[0].strip()
)
```

```python
df1['period'] = df['availability'].apply(
    lambda x: (
        re.findall(r'(?:since|from)\s+(.*?)(?:\*|$)', str(x))[0]
        if re.findall(r'(?:since|from)\s+(.*?)(?:\*|$)', str(x))
        else (
            re.findall(r'\((.*?)\)', str(x))[0]
            if re.findall(r'\((.*?)\)', str(x))
            else None
        )
    )
)
```

```python
df1['start_period'] = df1['period'].apply(
    lambda x: re.findall(r'^(.*?)\s*(?:-|$)', str(x))[0].strip()
)
```

```python
df1['end_period'] = df1['period'].apply(
    lambda x: (
        re.findall(r'-\s*(.*)$', str(x))[0]
        if re.findall(r'-\s*(.*)$', str(x))
        else 'Present'
    )
)
```

```python
df1["drive_type"] = df["drive"].apply(
    lambda x: re.findall(r"^\S+", str(x))[0]
).astype(str)
```

```python
df1["market_segment"] = df["market_segment"]
```

```python
df1["seats"] = df["seats"]
```

```python
df["heat_pump"].unique()
```

```python
def heat_pump(pump):
  if pump == "Heat pump available":
    return "Yes"
  else:
    return "No"

```

```python
df1["heat_pump_availability"] = df["heat_pump"].apply(heat_pump)
```

```python
df["towbar"].unique()
```

```python
def tow_bar(bar):
  if bar == "Towbar possible":
    return "Yes"
  else:
    return "No"

```

```python
df1["towbar_possible"] = df["towbar"].apply(tow_bar)
```

```python
df["safety_rating"].unique()
```

```python
def safety_rating(rating):
  if rating < 0:
    return 0
  else:
    return rating

```

```python
df1["safety_rating"] = df["safety_rating"].apply(safety_rating)
```
