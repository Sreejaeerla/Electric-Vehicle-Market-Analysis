
# Web Scraping and Exploratory Data Analysis of Electric Vehicle Specifications and Market Trends in Europe

# Problem Statement
The European electric vehicle market has grown rapidly, offering consumers a broad selection of vehicles with diverse capabilities, features, and performance levels. This increasing variety has made it difficult to assess how vehicles differ in terms of practicality, value, and overall market positioning. In addition, the competitive nature of the industry requires manufacturers to understand market dynamics, customer preferences, and emerging trends to maintain a strong presence. Therefore, there is a need to systematically examine the current EV landscape to identify patterns in vehicle offerings, brand presence, feature availability, and market distribution, enabling a better understanding of the factors shaping the European electric vehicle industry.


# Objective 
The objective of this project is to extract, clean, and analyze electric vehicle data collected from the EV Database using web scraping and data preprocessing techniques. The study aims to examine vehicle characteristics, manufacturer participation, availability trends, pricing patterns, and feature adoption across the European market. Through exploratory data analysis and visualization, the project seeks to uncover meaningful patterns, identify the key factors influencing vehicle value, compare different market categories, and generate actionable insights that support informed decision-making for consumers, manufacturers, and industry stakeholders.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import warnings
warnings.filterwarnings('ignore')


name = []

efficiency = []
real_range = []
weight = []
acceleration = []
one_stop_range = []
fast_charge = []
tow_weight = []
cargo_volume = []
price_per_range = []
germany_price_euros = []
netherlands_price_euros = []
uk_gbp = []
availability = []

drive = []
market_segment = []
seats = []
heat_pump = []
towbar = []
safety_rating = []

total_time = time.time()

for i in range(137):
    start_time = time.time()
    
    url = f'https://ev-database.org/#group=vehicle-group&rs-pr=10000_100000&rs-er=0_1000&rs-ld=0_1000&rs-ac=2_23&rs-dcfc=0_400&rs-ub=10_200&rs-tw=0_3000&rs-ef=100_350&rs-sa=-1_5&rs-w=1000_3500&rs-c=0_5000&rs-y=2010_2030&s=1&p={i}-10'
    response = requests.get(url)
    pagecontent = response.text
    soup = BeautifulSoup(pagecontent)
    # name 
    for j in soup.find_all('a',class_ = 'title'):
        name.append(j.find('span',class_= 'hidden').text)
        
    for l in soup.find_all('div',class_ = 'specs'):
        efficiency.append(l.find('span',class_ = 'efficiency').text)
        real_range.append(l.find('span',class_ = 'erange_real').text)
        weight.append(l.find('span',class_ = 'weight_p').text)
        acceleration.append(l.find('span',class_ = 'acceleration_p').text)
        one_stop_range.append(l.find('span',class_ = 'long_distance_total').text)
        fast_charge.append(l.find('span',class_ = 'battery_p').text)
        tow_weight.append(l.find('span',class_ = 'towweight_p').text)
        cargo_volume.append(l.find('span',class_ = 'cargo').text)
        price_per_range.append(l.find('span',class_ = 'priceperrange_p').text)

    for m in soup.find_all('div',class_ = 'pricing org'):
        germany_price_euros.append(m.find('span',class_ = 'country_de').text)
        netherlands_price_euros.append(m.find('span',class_ = 'country_nl').text)
        uk_gbp.append(m.find('span',class_ = 'country_uk').text)
  
    for n in soup.find_all('div', attrs = {'data-tooltip':'Date vehicle was available to order / is expected to become available to order.'}):
        availability.append(n.text.strip())
        
    for o in soup.find_all('div',class_ = 'icons-row-1'):
        seats.append(o.find('span',class_ = 'tooltip-wrapper').text.strip())
        drive.append((o.find('span',attrs = {'data-tooltip': lambda x: x and 'Drive' in x}))['data-tooltip'])
    
    for p in soup.find_all('span', attrs = {'data-tooltip': lambda x: x and 'Heat pump' in x}) :
        heat_pump.append(p['data-tooltip'])

    for q in soup.find_all('span', attrs = {'data-tooltip':'Market Segment'}) :
        market_segment.append(q.text)

    for r in soup.find_all('span', attrs = {'data-tooltip': lambda x: x and 'Towbar' in x}) :
        towbar.append(r['data-tooltip'])

    for s in soup.find_all('span', attrs = {'class':'safety hidden'}):
        safety_rating.append(s.text)
        
    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time))
    
```

```python
df = pd.DataFrame({
    'name':name,
    'efficiency':efficiency,
    'real_range':real_range,
    'weight':weight,
    'acceleration':acceleration,
    'one_stop_range': one_stop_range ,    
    'fast_charge': fast_charge ,
    'tow_weight': tow_weight ,
    'cargo_volume': cargo_volume ,
    'price_per_range ': price_per_range  ,
    'germany_price_euros': germany_price_euros ,
    'netherlands_price_euros':netherlands_price_euros  ,
    'uk_gbp':uk_gbp  ,
    'availability':availability  ,
    'drive':drive,
    'market_segment': market_segment ,
    'seats': seats ,
    'heat_pump':heat_pump  ,
    'towbar': towbar ,
    'safety_rating':safety_rating})
    
```
